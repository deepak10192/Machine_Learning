```python
import folium
```

```python
import pandas as pd
```

```python
dataset = pd.read_excel("covid_19_test_lab.xlsx")
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
hpmap = folium.Map(location=[20.316,82.235], zoom_start=10)
```

```python
dataset.columns
```

```python
place = dataset[['Latitude', 'Longitude']]
```

```python
place = place.values.tolist()
```

```python
i=0
for point in place:
    folium.Marker(location=point , 
                  popup=dataset['Test_lab'][i],
                 icon=folium.Icon(color='darkblue',icon='tint',icon_color='red')
                 ).add_to(hpmap)
    i+=1
```
